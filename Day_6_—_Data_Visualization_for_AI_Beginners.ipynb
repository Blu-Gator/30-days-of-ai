```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# nicer visual style
sns.set(style="whitegrid")
```

Matplotlib = core plotting library
Seaborn = prettier visuals on top of matplotlib

Use the same dataset from Day 5 to reinforce continuity.

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df.head()
```

Histograms show how values are distributed.

```python
plt.figure(figsize=(6,4))
df["sepal_length"].hist(color='skyblue')
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.show()
```

What to look for:

- Is the distribution wide or narrow?

- Are most values clustered?

- Are there long tails?

Mini task:

Make a histogram for petal_length using a different color.

```python
plt.figure(figsize=(6,4))
df["petal_length"].hist(color='red')
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Count")
plt.show()
```

Scatter plots reveal relationships between two variables.

```python
plt.figure(figsize=(6,4))
plt.scatter(df["sepal_length"], df["sepal_width"], alpha=0.6)
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

What to observe:

- Any upward/downward trend?

- Are the points clustered?

- Any obvious patterns?

Boxplots help visually compare groups or detect outliers.

Example: petal_length by species

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal Length Distribution by Species")
plt.show()
```

Interpretation:

- Where are the medians?

- Which species has longest petals?

- Any outliers?

Heatmaps help you see which features are strongly related.

Step 1: Create correlation matrix

```python
corr = df.select_dtypes(include=['float64', 'int64']).corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Interpretation:

- Values closer to 1 = strong positive correlation

- Values closer to -1 = strong negative correlation

- Near 0 = weak relationship
