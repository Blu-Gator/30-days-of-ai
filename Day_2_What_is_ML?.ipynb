What is a model?
  - A model is the result of training: you feed an algorithm lots of example data, and the learning process produces a function that maps inputs (like an image or a sentence) to outputs (like “cat” vs “dog” or a predicted number)
  - Once trained, that model can take new data it has never seen before and still output a useful prediction or classification, such as detecting spam emails or forecasting sales

What is data?
  - In everyday terms, data is any recorded fact or measurement, like a list of temperatures, customer purchases, or photos.

  - In machine learning, each data point is one example (like one image or one row in a table) and each feature is one measurable property of that example (like height, age, or pixel values).

  - Good data is usually organized, cleaned, and labeled (when needed) so that models can learn meaningful patterns instead of noise.

What does "training" mean?
  - In machine learning, “training” is the process of teaching a model to do a task by showing it lots of example data and letting it adjust itself to reduce mistakes.
